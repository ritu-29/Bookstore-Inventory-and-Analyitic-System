```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class Bookstore:
    def __init__(self):
        self.inventory = None
        self.sales = None
        self.last_plot = None 

    def load_inventory(self):
        try:
            path = input("Enter inventory CSV file path: ").strip().strip('"')
            self.inventory = pd.read_csv(path)
            print("Inventory dataset loaded successfully")
        except Exception as e:
            print("Inventory loading failed:", e)

    def load_sales(self):
        try:
            path = input("Enter sales CSV file path: ").strip().strip('"')
            self.sales = pd.read_csv(path)
            self.sales['Date'] = pd.to_datetime(self.sales['Date'])
            print("Sales dataset loaded successfully")
        except Exception as e:
            print("Sales loading failed:", e)
    def clean_data(self):
        print("\nChecking for null values...")

        if self.inventory is not None:
            print("\nInventory null values:")
            print(self.inventory.isnull().sum())
        else:
            print("Inventory not loaded")

        if self.sales is not None:
            print("\nSales null values:")
            print(self.sales.isnull().sum())
        else:
            print("Sales not loaded")


    def add_book(self):
        if self.inventory is None:
            self.inventory = pd.DataFrame(columns=["Title", "Author", "Genre", "Price", "Quantity"])
            print("Inventory initialized")

        title = input("Enter title: ")
        author = input("Enter author: ")
        genre = input("Enter genre: ")

        try:
            price = float(input("Enter price: "))
            quantity = int(input("Enter quantity: "))
        except ValueError:
            print("Price must be a number and quantity must be an integer")
            return

        if price <= 0 or quantity <= 0:
            print("Price and quantity must be positive")
            return

        new_book = {"Title": title, "Author": author, "Genre": genre, "Price": price, "Quantity": quantity}
        self.inventory = pd.concat([self.inventory, pd.DataFrame([new_book])], ignore_index=True)
        print("Book added successfully")

    def update_inventory(self):
        if self.inventory is None:
            print("Inventory not loaded")
            return
        title = input("Enter book title: ")
        if title not in self.inventory['Title'].values:
            print("Book not found")
            return
        try:
            quantity = int(input("Enter new quantity: "))
        except ValueError:
            print("Quantity must be an integer")
            return
        self.inventory.loc[self.inventory['Title'] == title, 'Quantity'] = quantity
        print("Inventory updated")

    def record_sale(self):
        if self.inventory is None or self.inventory.empty:
            print("Inventory not loaded")
            return
        title = input("Enter book title: ")
        if title not in self.inventory['Title'].values:
            print("Book not found")
            return
        try:
            quantity = int(input("Enter quantity sold: "))
        except ValueError:
            print("Quantity must be an integer")
            return

        stock = self.inventory.loc[self.inventory['Title'] == title, 'Quantity'].values[0]
        price = self.inventory.loc[self.inventory['Title'] == title, 'Price'].values[0]
        if quantity > stock:
            print("Not enough stock")
            return
            
        self.inventory.loc[self.inventory['Title'] == title, 'Quantity'] -= quantity
        new_sale = {"Date": pd.Timestamp.today(), "Title": title, "Quantity Sold": quantity, "Total Revenue": price * quantity}

        if self.sales is None:
            self.sales = pd.DataFrame([new_sale])
        else:
            self.sales = pd.concat([self.sales, pd.DataFrame([new_sale])], ignore_index=True)
        print("Sale recorded")

    def generate_report(self):
        total_revenue = self.sales['Total Revenue'].sum()
        best_seller = self.sales.groupby('Title')['Quantity Sold'].sum().idxmax()
        print("\nREPORT")
        print("Total Revenue:", total_revenue)
        print("Best Selling Book:", best_seller)

    def bar_chart(self):
        merged = self.sales.merge(self.inventory, on="Title", how="left")
        merged.groupby("Genre")["Total Revenue"].sum().plot(kind="bar")
        plt.title("Total Sales by Genre")
        self.last_plot = plt.gcf()
        plt.show()

    def line_chart(self):
        self.sales['Month'] = self.sales['Date'].dt.to_period('M')
        self.sales.groupby("Month")["Total Revenue"].sum().plot(kind="line")
        plt.title("Monthly Sales Trend")
        self.last_plot = plt.gcf()
        plt.show()

    def pie_chart(self):
        data = self.sales.groupby("Title")["Total Revenue"].sum().head(10)
        plt.figure()
        plt.pie(data, autopct="%1.1f%%")
        plt.title("Revenue Share (Top 10 Books)")
        plt.ylabel("")
        self.last_plot = plt.gcf()
        plt.show()

    def heatmap(self):
        merged = self.sales.merge(self.inventory, on="Title", how="left")
        corr = merged[['Price', 'Quantity Sold']].corr()
        sns.heatmap(corr, annot=True, cmap="viridis")
        plt.title("Price vs Quantity Sold Correlation")
        self.last_plot = plt.gcf()
        plt.show()

    def subplots(self):
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        fig.tight_layout(pad=5)

        # --- Bar Chart ---
        merged = self.sales.merge(self.inventory, on="Title", how="left")
        bar_data = merged.groupby("Genre")["Total Revenue"].sum()
        bar_data.plot(kind="bar", ax=axes[0, 0], color='skyblue')
        axes[0, 0].set_title("Total Sales by Genre")

        # --- Line Chart ---
        self.sales['Month'] = self.sales['Date'].dt.to_period('M')
        line_data = self.sales.groupby("Month")["Total Revenue"].sum()
        line_data.plot(kind="line", ax=axes[0, 1], marker='o', color='green')
        axes[0, 1].set_title("Monthly Sales Trend")

        # --- Pie Chart ---
        pie_data = self.sales.groupby("Title")["Total Revenue"].sum().head(10)
        pie_data.plot(kind="pie", autopct="%1.1f%%", ax=axes[1, 0])
        axes[1, 0].set_title("Revenue Share (Top 10 Books)")
        axes[1, 0].set_ylabel("") 

        # --- Heatmap ---
        corr = merged[['Price', 'Quantity Sold']].corr()
        sns.heatmap(corr, annot=True, cmap="coolwarm", ax=axes[1, 1])
        axes[1, 1].set_title("Price vs Quantity Sold Correlation")
        plt.show()



    def save_visualization(self):
        if self.last_plot is None:
            print("No visualization to save!")
            return

        folder = r"C:\Users\Ritu\Desktop\DATA_ANAYLIST"
        name = input("Enter file name (.png): ")
        full_path = f"{folder}\\{name}"

        self.last_plot.savefig(full_path)
        print(f"Visualization saved as {full_path} successfully!")



def menu():
    store = Bookstore()
    while True:
        print("\nBOOKSTORE INVENTORY")
        print("1. Load Inventory Dataset")
        print("2. Load Sales Dataset")
        print("3. Add Book")
        print("4. Show Null Values")        
        print("5. Update Inventory")
        print("6. Record Sale")
        print("7. Generate Report")
        print("8. Bar Chart")
        print("9. Line Chart")
        print("10. Pie Chart")
        print("11. Heatmap")
        print("12. Subplots")
        print("13. save Visualization")         
        print("14. Exit")                

        try:
            ch = int(input("Enter your choice: "))
        except ValueError:
            print("Invalid input, enter a number")
            continue

        if ch == 1:
            store.load_inventory()
        elif ch == 2:
            store.load_sales()
        elif ch == 3:
            store.add_book()
        elif ch == 4:
            store.clean_data()       
        elif ch == 5:
            store.update_inventory()
        elif ch == 6:
            store.record_sale()
        elif ch == 7:
            store.generate_report()
        elif ch == 8:
            store.bar_chart()
        elif ch == 9:
            store.line_chart()
        elif ch == 10:
            store.pie_chart()
        elif ch == 11:
            store.heatmap()
        elif ch == 12:
            store.subplots()
        elif ch == 13:
            store.save_visualization()     
        elif ch == 14:
            print("Exiting program")
            break
        else:
            print("Invalid choice")

if __name__ == "__main__":
    menu()
```
